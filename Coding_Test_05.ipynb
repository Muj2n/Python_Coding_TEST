```python
# 재귀함수 - 반복문에 대체됨(for문의 중첩)
def DFS(x): # 지역변수
  if x > 0:
    DFS(x-1) # Stack > X : 3, 2, 1로 쌓임
    print(x, end = '') # 1, 2, 3 호출

# 전역변수
n = int(input())
DFS(n)
```

```python
# 이진트리 순회(깊이우선탐색 : DFS)
# 깊이우선탐색 > 부모노드에서 왼쪽으로 끝까지 탐색

# 전위순회 : 부 > 왼 > 오
# 중위순회 : 왼 > 부 > 오
# 후위순회 : 왼 > 오 > 부

# 1. 전위순회 > 자기의 값을 호출 이후, 왼쪽노드와 오른쪽노드를 호출
def DFS(v):
  if v > 7:
    return # 7보다 커지면 함수 종료
  else:
    print(v, end = ' ') 
    DFS(v * 2) # 왼쪽노드 호출
    DFS(v * 2 + 1) # 오른쪽노드 호출

DFS(1)
```

```python
# 2. 중위순회 > 왼쪽노드 처리 이후 부모노드 마지막으로 오른쪽노드 처리

def DFS(v):
  if v > 7:
    return # 7보다 커지면 함수 종료
  else:
    DFS(v * 2) # 왼쪽노드 호출
    print(v, end = ' ') 
    DFS(v * 2 + 1) # 오른쪽노드 호출

DFS(1)
```

```python
# 3. 후위순회 > 왼쪽노드 처리 이후 오른쪽노드 처리 마지막으로 부모노드 처리
# 병합정렬에 사용
def DFS(v):
  if v > 7:
    return # 7보다 커지면 함수 종료
  else:
    DFS(v * 2) # 왼쪽노드 호출 
    DFS(v * 2 + 1) # 오른쪽노드 호출
    print(v, end = ' ')
# print가 되지 않는 순서에서 계속 밑의 노드로 내려감
DFS(1)
```

* 전역변수
  * 전체의 코드에서 해당되는 변수
* 지역변수
  * 함수나 해당되는 부분에서 해당되는 변수

```python
def DFS1():
  cnt = 3 # 지역변수
  print(cnt)

def DFS2():
  if cnt == 5: # 전역변수
    print(cnt)

def DFS3():
  global cnt # 전역변수로 해당
  if cnt == 5:
    cnt += 1 # 지역변수에 대한 값이라 판단 따라서 cnt값이 존재 X
    print(cnt)

cnt = 5 # 전역변수
DFS1()
DFS2()
DFS3()
print(cnt)
```

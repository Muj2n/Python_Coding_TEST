```python
# 이분검색 - SOL
# lt / rt
# mid = (lt + rt) // 2 
# 시간 복잡도 : logn

n, m = map(int, input().split())
a = list(map(int, input().split()))

# 이분 검색은 무조건 '정렬'이 기본
a.sort()
lt = 0
rt = n - 1

while lt <= rt:
  mid = (lt + rt) // 2
  if a[mid] == m:
    print(mid + 1) # mid : 인덱스 값
    break
  elif a[mid] >= m:
    rt = mid - 1
  else:
    lt = mid + 1
```

```python
# 랜선자르기(결정 알고리즘) - 이분 검색 : 답 > 특정 범위안에 있다는 것을 바로 알 수 있음
# 이분 검색 함수 만들기
# 1 ~ 가장 큰랜선 사이의 값

k, m = map(int, input().split())
# 함수 - 자른 전체 개수
def cnt(len):
  cnt = 0
  for x in line:
    cnt += (x // len)
  return cnt
line = []
res = 0 # 최대값
largest = 0
# 줄바꿔서 입력이 들어왔기 떄문 > input()
for i in range(k):
  tmp = int(input())
  line.append(tmp)
  largest = max(largest, tmp)
# line.sort()
lt = 1
# rt = line[-1]
rt = largest
while lt <= rt:
  mid = (lt + rt) // 2
  if cnt(mid) >= m:
    res = mid
    lt = mid + 1
  else:
    rt = mid - 1
print(res)
```

```python
# 뮤직비디오(결정 알고리즘)

# 함수 - 곡의 길이
def Count(cap):
  cnt = 1
  sum = 0
  for x in a:
    if sum + x > cap: # 음악의 용량을 초과
     cnt += 1
     sum = x
    else:
      sum += x
  return cnt

n, m = map(int, input().split())
a = list(map(int, input().split()))
res = 0
lt = 1
# 노래를 다 합한 값 : rt
rt = sum(a)
while lt <= rt:
  mid = (lt + rt) // 2
  if Count(mid) <= m:
    res = mid
    rt = mid - 1
  else:
    lt = mid + 1
print(res)
```

```python
# 마구간 정하기(결정 알고리즘)

# 함수 - 말의 마리수
def Count(len):
  cnt = 1
  ep = Line[0] # 첫번째 말 배치
  for i in range(1, n):
    if Line[i] - ep >= len: # 첫번째 말과 두번째 말의 거리
      cnt += 1
      ep = Line[i]
  return cnt
n, m = map(int, input().split())
Line = []
for _ in range(n):
  a = int(input())
  Line.append(a)
Line.sort()
res = 0
lt = 1 # 가능한 최소거리
rt = Line[n-1] # 가능한 최대거리
while lt <= rt:
  mid = (lt + rt) // 2 # 두 말의 최대 거리에서 줄여간다
  if Count(mid) >= m:
    res = mid
    lt = mid + 1
  else:
    rt = mid - 1
print(res)
```

```python
# 회의실 배정 (그리디 : 단계에서 가장 좋은 조건을 선택)
# 1. 정렬 > 선택

# 끝나는 시간으로 정렬

n = int(input())
meeting = []
for i in range(n):
  s, e = map(int, input().split()) # s : 시작시간 / e : 끝나는 시간
  meeting.append((s ,e)) # 튜플형태로 입력
meeting.sort(key = lambda x : (x[1], x[0])) # 정렬 순서 : 배열의 두번째 열의 오름차순 / 첫번째 열의 오름차순 

et = 0
cnt = 0
for s, e in meeting:
  if s >= et:
    et = e
    cnt += 1
print(cnt)
```

```python
# 씨름선수 (그리디)
# 키에 대한 정렬
# 몸무게를 MAX 이후 변경될 때 마다 +1

n = int(input())
body = []
for i in range(n):
  a, b = map(int, input().split())
  body.append((a, b))
body.sort(reverse = True) # 키의 내림차순 정렬
largest = 0
cnt = 0
for x, y in body:
  if y > largest:
    largest = y
    cnt += 1
print(cnt)
```

```python
# 창고정리 (그리디) - SOL
# sort를 매번 시행

l = int(input())
a = list(map(int, input().split()))
m = int(input())

a.sort()
for _ in range(m):
  a[0] += 1
  a[l - 1] -= 1
  a.sort()

print(a[l-1]- a[0])
```

```python
# 침몰하는 타이타닉 (그리디)

# 제일 가벼운사람과 제일 무거운사람을 같이 태워 보낸다 > 불가능하다면 제일 무거운 사람만 태우고 가야함

n, limit = map(int, input().split())
p = list(map(int, input().split()))
p.sort()
cnt = 0
while p: # p의 리스트가 존재하면 참 / 아니면 거짓
  if len(p) == 1:
    cnt += 1
    break
  if p[0] + p[-1] > limit:
    p.pop() # 가장 뒤를 pop
    cnt += 1
  else:
    p.pop(0) # 제일 앞의 자료가 사라짐
    p.pop()
    cnt += 1
# 마지막에 1명이 남는 경우 if문 위배됨 p[0] = p[-1]

print(cnt)
```

```python
# 침몰하는 타이타닉(그리디) - deque 이용 : 앞 / 뒤로 제거 가능, 리스트 : 앞에 문자 제거되면 한개씩 당겨짐
from collections import deque
n, limit = map(int, input().split())
p = list(map(int, input().split()))
p.sort()
cnt = 0
p = deque(p)
while p: # p의 리스트가 존재하면 참 / 아니면 거짓
  if len(p) == 1:
    cnt += 1
    break
  if p[0] + p[-1] > limit:
    p.pop() # 가장 뒤를 pop
    cnt += 1
  else:
    p.popleft() # 제일 앞의 자료가 사라짐 - 데크 자료구조에서 제일 앞자료 제거
    p.pop()
    cnt += 1
# 마지막에 1명이 남는 경우 if문 위배됨 p[0] = p[-1]

print(cnt)
```

```python
# 증가수열 만들기(그리디)

# lt / rt 만들기
# last : 수열의 마지막 값

n = int(input())
a =  list(map(int, input().split()))
lt = 0 # 제일 앞 순번
rt = n-1 # 제일 뒤 순번
last = 0 # 수열의 마지막 값
res = '' # 문자열 - 출력값
tmp = [] # 두개를 비교할 리스트

while lt <= rt:
  if a[lt] > last:
    tmp.append((a[lt], 'L'))
  if a[rt] > last:
    tmp.append((a[rt], 'R'))
  tmp.sort() # lt와 rt값의 비교 - 작은 값 들고오기
  if len(tmp) == 0: # 두 값이 모두 마지막 값보다 작을때 종료
    break
  else:
    res = res + tmp[0][1] # 리스트의 0번자료의 1번째 값
    last = tmp[0][0] # 리스트의 0번자료의 0번째 값
    if tmp[0][1] == 'L': # lt에서 들어간 값
      lt += 1
    else:
      rt += -1
  tmp.clear # 리스트값 초기화
print(len(res))
print(res)
```

```python
# 역수열 (그리디)

# 0의 값의 리스트 생성
# 0의 값이 있을 때, cnt += 1

n = int(input())
a =  list(map(int, input().split()))

seq = [0] * n # 새로운 리스트 0 ~ n-1 인덱스
for i in range(n):
  for j in range(n): # 빈공간 찾기 / 자기자리 찾기
    if a[i] == 0 and seq[j] == 0:
      seq[j] = i + 1 # i + 1 : 원래 본인의 값 
      break
    elif seq[j] == 0:
      a[i] -= 1 # 0의 자리를 발견했을 때 값이 한개씩 작아짐 - 다른 숫자의 공간
for x in seq:
  print(x, end = ' ')
```

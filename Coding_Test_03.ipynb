```python
# 이분검색

n, m = map(int, input().split())
a = list(map(int, input().split()))
a.sort()
for idx, value in enumerate(a):
  if value == m:
    print(idx+1) # 0부터 시작
```

```python
# 랜선자르기(결정 알고리즘)

n, k = map(int, input().split()) # n개의 줄 / k개 동일한 줄 만들기
a = int(input())
b = int(input())
c = int(input())
d = int(input())
res = 0
l = []
l.append(a)
l.append(b)
l.append(c)
l.append(d)
l.sort()
for x in l:
  cnt = 0
  for i in range(l[0],1,-1):
    cnt += x // i
if cnt >= k:
  print(cnt)
  

```

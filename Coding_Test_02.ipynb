```python
# 회문 문자열 검사 - SOL (1)

n = int(input())
for i in range(n):
  s = input() # 문자열 한개가 들어옴
  s = s.upper() # 대소문자 구별 X
  size = s.len(s) # 문자열의 길이
  for j in range(size // 2): # 홀수 시, 가운데 제외하고 앞과 뒤 비교
    if s[j] != s[-j-1]:
      print('#%d NO' %(i + 1))
      break # for문 중단
  else: # for - else구문
    print('#%d YES' %(i + 1))
```

```python
# 숫자만 추출 - SOL
s = input() # 문자열은 그대로 받아도됨
res = 0
cnt = 0
for x in s:
  if x.isdecimal(): # isdigit() : 숫자형태는 다찾아줌 / isdecimal() : 0 ~ 9까지 받아줌
    res = res * 10 + int(x) # 숫자 만들기 : 자리수 맞추기
  for i in range(1, res + 1): # 약수 구하기
    if res % i == 0:
      cnt += 1
print(cnt)
```

```python
# 카드 역배치 - SOL
# a, b = b, a > a, b의 자리 바꿈

a = list(range(21)) # 1 ~ 20 리스트 출력

for _ in range(10): # _ : 변수 X > 10개의 구간
  s, e = map(int, input().split())
  # (e - s + 1) // 2 변경 횟수
  for i in range((e - s + 1) // 2):
    a[s + i], a[e - i] = a[e - i], a[s + i] # 맨 앞의 값과 맨 뒤의 값 변경
a.pop(0) # pop : 0번 index 지우기
for x in a:
  print(x, end = ' ')
```

```python
# 두 리스트 합치기 - SOL

# 정렬 사용 시 : nlogn > 숫자 비교 : n

# p1 / p2 비교 후 작은 것 대입 > 대입하고 증가

n = int(input())
a = list(map(int, input().split()))
m = int(input())
b = list(map(int, input().split()))

p1 = p2 = 0 # 포인트 변수 초기화
c = [] # 빈 리스트
while p1 < n and p2 < m: # 포인트가 끝까지 가면 멈춤
  if a[p1] <= b[p2]: # 포인터의 해당 값이 작으면
    c.append(a[p1]) # 빈 리스트에 추가
    p1 += 1 # 포인터 1 증가
  else:
    c.append(b[p2])
    p2 += 1
# 누군가 하나가 처리가 완료 된 후
if p1 < n: # 이후 값들 모두 추가
  c = c + a[p1:]
else:
  c = c + b[p2:]
for x in c:
  print(x, end = ' ')
```

```python
# 수들의 합 - SOL

# 부분적인 연속의 합

n, m = map(int, input().split())

a = list(map(int, input().split()))

lt = 0 # 왼쪽 인덱스 
rt = 1  # 오른쪽 인덱스
tot = a[0] # 리스트의 lt ~ rt전까지의 합
cnt = 0 # 개수

# while문 사용
while True:
  if tot < m: # 1) tot 값이 m 보다 작을 때
    if rt < n: # rt의 인덱스 값의 증가
      tot += a[rt] 
      rt += 1 # 1증가
    else: # rt의 인덱스 값이 열의 마지막 값(n)과 같으면 break
        break
  elif tot == m: # 2) tot 값이 m과 같을 때
    cnt += 1
    tot -= a[lt] # 합계 값에 현재 lt의 위치의 있는 값 삭제
    lt += 1 # lt 위치 +1
  else: # 3) tot 값이 m 보다 클때
    tot -= a[lt] # 합계 값에 현재 lt의 값 삭제
    lt += 1 # lt 위치 + 1
print(cnt)
```

```python
# 격자판 최대합 - SOL
# 2차원 리스트 생성

n = int(input())

a = [list(map(int, input().split())) for _ in range(n)] # 한줄이 리스트화 / for _ in range(n) : 변수 없이 n행 읽게해줌
# 행렬처럼 나타내기
# for x in a:
largest = -2147000000
for i in range(n):
  sum1 = sum2 = 0 
  for j in range(n):
    sum1 += a[i][j] # 행의 합 : sum1
    sum2 += a[j][i] # 열의 합 : sum2
  if sum1 > largest: 
    largest = sum1
  if sum2 > largest: 
    largest = sum1
# 대각선 합
sum1 = sum2 = 0
for i in range(n):
  sum1 += a[i][i] # \ 대각선
  sum2 += a[i][n-i-1] # / 대각선
if sum1 > largest: 
  largest = sum1
if sum2 > largest: 
  largest = sum1

print(largest)
```

```python
# 사과나무(다이아몬드) - SOL
n = int(input())

a = [list(map(int, input().split())) for _ in range(n)]
res = 0
# s = e = n // 2
# s -= 1
# e += 1 
# 중간지점 이후 감소 진행
# s += 1
# e -= 1

# 다이야몬드 더하기
s = e = n // 2
for i in range(n):
  for j in range(s, e+1):
    res += a[i][j] # i 행 / j 열
  if s < n // 2: # 중간지점보다 작아질때
    s -= 1
    e += 1
  else:
    s += 1
    e -= 1
print(res)
```

```python
# 곶감(모래시계)
n = int(input())

a = [list(map(int, input().split())) for _ in range(n)]

m = int(input())
for i in range(m):
  h, t, k = map(int, input().split())
  if t == 0: # 왼쪽방향
    for _ in range(k): # k번 뽑기 : pop(0) > 제일 참 / pop() > 제일 뒤
      a[h-1].append(a[h-1].pop(0)) # 0번째 값 뽑기
  else: # 오른방향
    for _ in range(k): # k번 뽑기 : pop
      a[h-1].insert(0, a[h-1].pop()) # 마지막번째 값 뽑기    
# 모래시계 더하기
res = 0
s = 0
e = n - 1
for i in range(n):
  for j in range(s, e + 1):
    res += a[i][j] # i 행 / j 열
  if s < n // 2: # 중간지점보다 작아질때
    s += 1
    e -= 1
  else:
    s -= 1
    e += 1
print(res)
```

```python
# 봉우리 - SOL

n = int(input())
dx = [-1, 0 , 1, 0] # 상하좌우 비교를 위해 행/열 생성
dy = [0, 1, 0, -1]  # 순서 : 12 > 3 > 6 > 9
a = [list(map(int, input().split())) for _ in range(n)]

# 가장자리 '0' 초기화
a.insert(0, [0]*n) # 1행에 '0'열 대입
a.insert(0, [0]*n) # 마지막행에 '0'열 대입
for x in a: # 리스트의 한행
  x.insert(0, 0) # 1열에 '0' 대입
  x.append(0) # 마지막열에 '0' 대입

# 상하좌우 비교
cnt = 0 # 개수 세기
for i in range(1, n + 1): # 0에 대해서 비교 X
  for j in range(1, n + 1):
      if all(a[i][j]> a[i+dx[k]][j + dy[k]]) for k in range(4)): # 모두가 참일 때, 참 / 상하좌우 비교(4가지 방향 탐색 시 사용)
      cnt += 1

print(cnt)
```

```python
# 스도쿠 검사  - SOL
# 체크리스트 만들기 : 행 / 열 / 그룹 > 3개 만들기

def check(a):
  for i in range(9):
    ch1 = [0] * 10 # 행 체크리스트
    ch2 = [0] * 10 # 열 체크리스트
    for j in range(9):
      ch1[a[i][j]] = 1 # 체크리스트에 1값 추가
      ch2[a[j][i]] = 1 # 체크리스트에 1값 추가
    if sum(ch1) != 9 or sum(ch2) != 9:
      return False # 체크리스트의 합 = 9 > 참 
# 그룹 체크리스트
  for i in range(3): # 그룹의 첫번째 : 열
    for j in range(3): # 그룹의 첫반째 : 행
      ch3 = [0]*10
      for k in range(3): # i에 더해지는 값(0, 1, 2)
        for s in range(3): # j에 더해지는 값(0, 1, 2)
          ch3[a[i*3+k][j*3+s]] = 1 # (3,3)그룹에서 (1, 1 > 2 > 3)
      if sum(ch3) != 9:
        return False
  return True
```
